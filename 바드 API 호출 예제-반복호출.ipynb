```python
from bardapi import Bard
import requests
import os

os.environ['_BARD_API_KEY'] = "you_api_key"
```

```python
session = requests.Session()
session.headers = {
            "Host": "bard.google.com",
            "X-Same-Domain": "1",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36",
            "Content-Type": "application/x-www-form-urlencoded;charset=UTF-8",
            "Origin": "https://bard.google.com",
            "Referer": "https://bard.google.com/",
        }
session.cookies.set("__Secure-1PSID", os.getenv("_BARD_API_KEY")) 

bard = Bard(token=os.getenv("_BARD_API_KEY"), session=session, timeout=30)
```

```python
result = bard.get_answer("다양한 에세이 주제 한 문장으로 10개 알려줘")['content']
print(result)
line = result.split('\n')
```

```python
result_title = []
result_list = []

for i in range(2, 12):
    result_title.append(line[i])
    print("질문 : "  + line[i])
    print("")
    result = bard.get_answer("'" + line[i] + "' 라는 주제로 에세이 써줘")['content']
    print("답변 : "  + result)
    print("")
    result_list.append(result)
```

```python
for i in range(0,10):
    title = result_title[i]
    content = result_list[i]
    
    with open(title.replace('?', '')+".txt", "w",encoding="utf-8") as f:
        f.truncate(0)
        f.write(content)
    f.close()
```
