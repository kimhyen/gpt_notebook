```python
service_key = "you api service_key"
from PyKakao import KoGPT
api = KoGPT(service_key = service_key)
```

```python
# 필수 파라미터
prompt = "인간처럼 생각하고, 행동하는 '지능'을 통해 인류가 이제까지 풀지 못했던"
max_tokens = 64

# 결과 조회
result = api.generate(prompt, max_tokens, temperature=0.7, top_p=0.8)
print(result)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
from PyKakao import Karlo

# Karlo API 인스턴스 생성
api = Karlo(service_key)

img_dict = api.text_to_image("cute dog and cat image", 1)

# 생성된 이미지 정보
img_str = img_dict.get("images")[0].get('image')

# base64 string을 이미지로 변환
img = api.string_to_image(base64_string = img_str, mode = 'RGBA')

plt.imshow(img)
plt.axis('off')  # 축 제거
plt.show()
```
